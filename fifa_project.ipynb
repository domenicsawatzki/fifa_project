```python
def clean_value(x):
     
    unit = x[-1] 
    
    if unit == 'M':
         x = float(x[:-1])
        
    elif unit == 'K':
        x = float(float(x[:-1]) / 1000)
    
    else: x = 0
    
    return x 

df_1['value'] = df_1['value'].str.replace('€', '')
df_1['value'] = list(map(clean_value, df_1['value']))
df_1['ir'] = df_1['ir'].replace('[^0-9]', '', regex=True)

df_1

```

```python
df_1['ir'] = pd.to_numeric(df_1['ir'])


print(df_1['hits'].unique())

def clean_hits(x):
     
    unit = x[-1] 
    
    if unit == 'K':
        x = float(x[:-1])
        x = x * 1000
    
    else: x     
    
    return x 


df_1['hits'] = list(map(clean_hits, df_1['hits']))
df_1['hits'].unique()

df_1['hits'] = pd.to_numeric(df_1['hits'])


```

```python
display(df_1)

y = df_1['ova']
X = df_1.drop(['ova'], axis=1)
```

```python
X_num = df_1.select_dtypes(include = np.number)
display(X_num)

X_cat = df_1.select_dtypes(include = object)
display(X_cat)
```

```python
from sklearn.preprocessing import MinMaxScaler

display(X_num.describe().T)

MinMaxtransformer = MinMaxScaler().fit(X_num)
X_normalized = MinMaxtransformer.transform(X_num)

X_normalized = pd.DataFrame(X_normalized,columns=X_num.columns)
display(X_normalized.head())


X_normalized.describe().T
```

```python

# pd.DataFrame(OneHotEncoder(drop='first').fit_transform(df_cat).toarray(),
#              columns=OneHotEncoder(drop='first').fit(df_cat).get_feature_names(input_features=df_cat.columns)).head()
```

```python
from sklearn.preprocessing import OneHotEncoder

# Initialize the OneHotEncoder
encoder = OneHotEncoder(drop='first')

# Fit the encoder on the categorical data
encoder.fit(X_cat)

# Transform the categorical data using the fitted encoder
encoded_X_cat = encoder.transform(X_cat)

# Convert the encoded data to a pandas DataFrame
encoded_X_cat = pd.DataFrame(encoded_X_cat.toarray(), columns=encoder.get_feature_names_out(X_cat.columns))

encoded_X_cat
```

```python
correlations_matrix = X_num.corr()
correlations_matrix = correlations_matrix.round(2)
correlations_matrix
```

```python
X = pd.concat([X_normalized ,encoded_X_cat], axis=1)

display(X)

X.to_csv('clean_df.csv', sep=',', encoding='utf-8', index=False)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression as LinReg

linreg=LinReg()    # model
linreg.fit(X_train, y_train)   # model train
```

```python

from sklearn.metrics import r2_score

train_r2=r2_score(linreg.predict(X_train), y_train)
test_r2=r2_score(linreg.predict(X_test), y_test)



print ('train r2 score: {} -- test r2 score: {}'.format(train_r2, test_r2))
```

```python
from sklearn.metrics import mean_squared_error as mse

train_mse=mse(linreg.predict(X_train), y_train)
test_mse=mse(linreg.predict(X_test), y_test)


print ('train MSE: {} -- test MSE: {}'.format(train_mse, test_mse))
print ('train RMSE: {} -- test RMSE: {}'.format(train_mse**.5, test_mse**.5))
```
